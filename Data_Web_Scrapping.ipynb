```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.flipkart.com/search?q=smartwatch&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
req=requests.get(url)
```

```python
content=BeautifulSoup(req.content,'html.parser')
```

```python
print(content)
```

```python
data=content.find_all('div',{'class':'_2B099V'})
```

```python
print(data)
```

```python
links=[]
watch_name=[]
start_link="https://www.flipkart.com/"
```

```python
for items in data:
  rest_links=items.find('a')['href']
  name=items.find('a',attrs={'class':'IRpwTa'})
  watch_name.append(name.text)
  links.append(start_link+rest_links)
```

```python
print(watch_name[0])
print(links[0])
```

```python


dataframe = {'watch_names': watch_name}
Final_dataframe = pd.DataFrame(dataframe)
print(Final_dataframe)
```

```python
Final_dataframe.to_csv('datasetscrapped.csv')
```
